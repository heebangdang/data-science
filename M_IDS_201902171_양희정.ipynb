```python
from scipy.stats import ttest_ind
ttest,pval = ttest_ind(data4['Temperature'],data4['Humidity'])
print("ttest", ttest)
print("p-value",pval)

if pval <0.05:
  print("we reject null hypothesis")
else:
  print("we fail to reject (accept) null hypothesis")
```

Null hypothesis: The mean of "Temperature" data equals to the mean of "Humidity" data
Alternate hypothesis: The mean of "Temperature" data does not equal to the mean of "Humidity" data
We reject null hypothesis, So the mean of "Temperature" does not equal to the mean of "Humidity" data
