```python
# (d)
plt.scatter(data['depth'], data['table']) 
plt.title("depth vs table")
plt.xlabel("Depth")
plt.ylabel("Table")
plt.show()
```

```python
# Problem 2
# (a)
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.heatmap(tips.corr(), annot=True)
```

This graph consists of a total of nine compartments, expressed in numbers from 0.49 to 1.

```python
# (b)
import seaborn as sns
sns.pairplot(tips, hue='day',diag_kws={'bw':'1.0'})
```

In total_bill and total_bill relationship, Fri represents the largest value, Sat represents the largest value in tip and total_bill relationship, and Sat represents the largest value in size and total_bill relationship. In the relationship between tip and total_bill, Sat represents the largest value, Fri represents the largest value in the relationship between tip and tip, and Sat represents the largest value in size and tip. Thur represents the largest value in total_bill and size relationships, Thur and Sun represent the largest values in tip and size relationships, and Fris represents the largest value in size and size relationships.

```python
# (c)
tips['tip_percent'] = 100 * tips['tip'] / tips['total_bill']
with sns.axes_style(style='ticks'):
    g = sns.catplot("day", "tip_percent", "time", data=tips, kind="box")
    g.set_axis_labels("Day", "Tip_percent");
```

It represents the highest tip_percent value in Lunch and the lowest tip_percent value in Dinner.There's a lot of lunch distribution on Thursdays, and a lot of dinners on Saturdays and Sundays. On Friday, Lunch and Dinner are distributed together.

```python
# (d)
grid = sns.FacetGrid(tips, row="time", col="size", margin_titles=True)
grid.map(plt.hist, "tip_percent");
```

When size is 2, and Dinner, it can be seen that it is more distributed than other values. And when it is Lunch and size is 2, it shows that it is more distributed than when it has values of other sizes.
