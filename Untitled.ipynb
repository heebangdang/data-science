```python
from sklearn.preprocessing import StandardScaler
features = ['call quality','phone','support service','supplementary service','support facility','cost']
```

```python
x = data.loc[:, features].values # array # feature에 있는 값 반환
# Separating out the target
y = data.loc[:,['service']].values # array # 'target' attribute에 있는 값 반환
```

```python
x = StandardScaler().fit_transform(x)
```

```python
from sklearn.decomposition import PCA # 분해(decomposition)
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])
```

```python
finalDf = pd.concat([principalDf, data[['service']]], axis = 1)
finalDf
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = ['자사', '경쟁1', '경쟁2','경쟁3']
colors = ['r', 'g', 'b','y'] # red green blue
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['service'] == target #'target'이름만 바꿔주기
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```

```python
from sklearn.datasets import load_digits
digits = load_digits()
digits.data.shape
```

```python
pca = PCA(2)  # project from 64 to 2 dimensions
projected = pca.fit_transform(digits.data)
print(digits.data.shape)
print(projected.shape)
```

```python
plt.scatter(projected[:, 0], projected[:, 1],
            c=digits.target, edgecolor='none', alpha=0.8,
            cmap=plt.cm.get_cmap('Spectral'))
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.colorbar();
```
