```python
# (c)
plt.hist(data['price'], bins = 10)
plt.xlabel("price")
plt.ylabel("variable")
plt.show()
```

The total number is 53940, which does not fall apart exactly, so the exact bin cannot be obtained. And the lower the price, the more variable the value is distributed.

```python
# (d)
plt.scatter(data['depth'], data['table']) 
plt.title("depth vs table")
plt.xlabel("Depth")
plt.ylabel("Table")
plt.show()
```

Depth is the height of a diamond, Table is the width of a diamond. The two values should not be too large or too small. Therefore, it can be seen that there is a large distribution in the middle value.

```python
# (e)
import seaborn as sns
sns.set()
```

```python
grid = sns.FacetGrid(data, row="cut", col="color", margin_titles=True)
grid.map(plt.hist, "carat");
```

It can be seen that the cut has the largest number of 'Ideal' values, and the type of color does not matter much.

```python
# (f)
with sns.axes_style(style='ticks'):
    g = sns.catplot("color", "carat", "cut", data=data, kind="box")
    g.set_axis_labels("Color", "Carat");
```

It has the highest carat in Color 'J'.

```python
# Problem 2
# (a)
tips = sns.load_dataset('tips')
```

```python
sns.heatmap(tips.corr(), annot=True)
```

This graph consists of a total of nine compartments, expressed in numbers from 0.49 to 1.

```python
# (b)
sns.pairplot(data=tips, hue="day", diag_kind='auto')
```

```python
# (c)
tips['tip_percent'] = 100 * tips['tip'] / tips['total_bill']
with sns.axes_style(style='ticks'):
    g = sns.catplot("day", "tip_percent", "time", data=tips, kind="box")
    g.set_axis_labels("Day", "Tip_percent");
```

```python
grid = sns.FacetGrid(tips, row="time", col="size", margin_titles=True)
grid.map(plt.hist, "tip_percent");
```

When size is 2, and Dinner, it can be seen that it is more distributed than other values. And when it is Lunch and size is 2, it shows that it is more distributed than when it has values of other sizes.
